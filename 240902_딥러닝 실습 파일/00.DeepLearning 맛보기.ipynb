```python
# 문제데이터 : studytime
# 정답데이터 : G3
# X 문제데이터 > 2차원 이상의 데이터 > 벡터로 변환 > 2차원 이상의 벡터는 대문자 표기
# y 정답데이터 > 1차원 데이터 > 벡터로 변환 > 1차원 벡터는 소문자 표기
X = data[['studytime']]
y = data['G3']
X.shape, y.shape
```

```python
X
```

```python
# 훈련과 평가 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,
                                                    random_state = 5)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 머신러닝으로 성적 예측해보기
# 선형회귀모델 가져오기
from sklearn.linear_model import LinearRegression
# 평가지표 > 평균제곱오차
from sklearn.metrics import mean_squared_error
# 모델 불러오기
lr = LinearRegression()
# 학습하기
lr.fit(X_train,y_train)
# 예측하기
y_pred = lr.predict(X_test)
# 평가하기
mean_squared_error(y_test,y_pred)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
# 딥러닝으로 성적 예측해보기
# 모델 설계하기 = MLP(Multi Layer Perceptron)
model = Sequential()
model.add(Dense(units = 32, input_dim = 1, activation = 'relu'))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dense(units = 32, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))

model.compile(
    loss = 'mean_squared_error',
    optimizer = 'adam',
    metrics = ['mse']
)
# 학습하기 > 머신러닝 : 한번 학습하고 끝, 딥러닝 : 여러번 학습 > 결과를 계속 개선시킴
model.fit(X_train,y_train,epochs = 100)

```

```python
# 예측하기
model.predict(pd.DataFrame([3]))
```

```python
# 평가하기
model.evaluate(X_test,y_test)
```
